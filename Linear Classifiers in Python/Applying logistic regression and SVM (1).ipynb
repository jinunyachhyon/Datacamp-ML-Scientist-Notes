# Applying logistic regression and SVM

```python
import sklearn.datasets

#load newsgroups data from scikit-learn's repository
newsgroups = sklearn.datasets.fetch_20newsgroups_vectorized()

X,y = newsgroups.data, newsgroups.target
```

```python
X.shape , y.shape
```

```python
#fitting and predicting
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X,y)

y_pred = knn.predict(X)
```

```python
# Model Evaluation
knn.score(X,y)
```

```python
# Model Evaluation on validation set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y)

knn.fit(X_train, y_train)
knn.score(X_test, y_test)
```

## Logistic Regression

```python
#Load wine dataset from sklearn library
import sklearn.datasets
wine = sklearn.datasets.load_wine()

# Import LogisticRegression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(wine.data,wine.target)

# Compute training accuracy
lr.score(wine.data, wine.target)
```

```python
# LogisticRegression can also output confidence sccores rather than "hard" or definite predictions

lr.predict_proba(wine.data[:1])
```

## LinearSVC (Linear Support Vector Classifier)
In scikit-learn, the basic SVM classifier is called LinearSVC.

```python
# Import LinearSVC
from sklearn.svm import LinearSVC

svm = LinearSVC()
svm.fit(wine.data,wine.target)

# Compute training accuracy
svm.score(wine.data, wine.target)
```

## SVC (Support Vector Classifier)
It fits a nonlinear SVM by default.

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(wine.data, wine.target)

# Compute training accuracy
svm.score(wine.data, wine.target)
```
