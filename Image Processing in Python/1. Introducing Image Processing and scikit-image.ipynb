# Getting started with thresholding

## Thresholding
- Partitioning an image into a foreground and background by making it **black and white.**
- We do so by setting each pixel to:
    - 255 (white) if pixel > thresh value
    - 0 (black) if pixel < thresh value

![image](image.png)


```python
# Basic application of thresholding
from skimage.color import rgb2gray
import matplotlib.pyplot as plt

# Load image
image = plt.imread('./pexels-andrew-neel-3143922.jpg')
plt.imshow(image)
```

```python
# Obtain optimal threshold value
threshold = 127

# Apply thresholding to the image
binary = image[:,:,0] > threshold

# Show the original and threshold
plt.imshow(binary)
```

## Categories 
- Global or histogram based: good for uniform backgrounds
- Local or adaptive: for uneven bakground illumination

![image-2](image-2.png)


## 1. Global Thresholding 

```python
# Global thresholding

# Import the otsu threshold function
from skimage.filters import threshold_otsu

# Obtain the optimal threshold value
thresh = threshold_otsu(image[:,:,0])

# Apply thresholding to the image
binary_global = image[:,:,0] > thresh

# Visualize 
plt.imshow(binary_global)
```

## 2. Local Thresholding

```python
# Import the local threshold function, this function calc thresholds in small pixel regions surrounding each pizel we are binarizing
from skimage.filters import threshold_local 

# Set the block size to 35 
block_size = 35

# Obtain the optimal local thresholding
local_thresh = threshold_local(image[:,:,0], block_size, offset=10)

# Apply local thresholding and obtain the binary image
binary_local = image[:,:,0] > local_thresh

# Visualize
plt.imshow(binary_local)
```
